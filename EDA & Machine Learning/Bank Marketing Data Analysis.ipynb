### Classifying Age in to Different Groups

```python
def transfer_age(x):
    if 17 <= x <= 30:
        return 'Young_Adults'
    elif 30 < x <= 40:
        return 'Adults'
    elif 40 < x <= 50:
        return 'Elders'
    else:
        return 'Older_Adults'
```

```python
df_vis['Age'] = df_vis['Age'].apply(transfer_age)
```

## Machine Learning Algorithms

```python
df_ml = df.copy()
```

```python
df_ml
```

### Replacing Characters to Numerics to Perform Models on Dataset

```python
df_ml.replace({'no':0,'yes':1},inplace=True)
df_ml.replace({'admin.':0,'blue-collar':1,'technician':2,'services':3,'management':4,'retired':5,'entrepreneur':6,
               'self-employed':7,'housemaid':8,'unemployed':9,'student':10},inplace=True)
df_ml.replace({'married':0,'single':1,'divorced':2},inplace=True)
df_ml.replace({'illiterate':0,'basic.4y':1,'basic.6y':2,'basic.9y':3,'high.school':4,'university.degree':5,
            'professional.course':6},inplace=True)
df_ml.replace({'mon':0,'tue':1,'wed':2,'thu':3,'fri':4},inplace=True)
df_ml.replace({'telephone':0,'cellular':1},inplace=True)
df_ml.replace({'nonexistent':0,'failure':1,'success':2},inplace=True)
df_ml.replace({'mar':0,'apr':1,'may':2,'jun':3,'jul':4,'aug':5,'sep':6,'oct':7,'nov':8,'dec':9},inplace=True)
```

```python
df_ml
```

### Assigning Train & Test Data to Perform Models on Dataset

```python
x = df_ml.drop(columns=['Status'],axis=1)
```

```python
x
```

```python
y = df_ml.Status
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

#### Observations:
* 70% of Data is assigned to Train Data
* 30% of Data is assigned to Test data

```python
from sklearn.metrics import classification_report
```

### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
model_LR = LogisticRegression()
```

```python
model_LR.fit(x_train,y_train)
```

```python
acc_LR = model_LR.score(x_test,y_test)*100
acc_LR
```

```python
y_predict_LR = model_LR.predict(x_test)
```

```python
print('LogisticRegression - Classification Report\n')
print(classification_report(y_test,y_predict_LR))
```

#### Observation:
* Accuracy of Data by applying Logistic Regression is 90%

### Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model_DT = DecisionTreeClassifier()
```

```python
model_DT.fit(x_train,y_train)
```

```python
acc_DT = model_DT.score(x_test,y_test)*100
acc_DT
```

```python
y_predict_DT = model_DT.predict(x_test)
```

```python
print('DecisionTreeClassifier - Classification Report\n')
print(classification_report(y_test,y_predict_DT))
```

#### Observations:
* Accuracy of Data by applying Decision Tree Classifier is 89%

### Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model_RF = RandomForestClassifier()
```

```python
model_RF.fit(x_train,y_train)
```

```python
acc_RF = model_RF.score(x_test,y_test)*100
acc_RF
```

```python
y_predict_RF = model_RF.predict(x_test)
```

```python
print('RandomForestClassifier - Classification Report\n')
print(classification_report(y_test,y_predict_RF))
```

#### Observations:
* Accuracy of Data by applying Random Forest Classifier is 92%
